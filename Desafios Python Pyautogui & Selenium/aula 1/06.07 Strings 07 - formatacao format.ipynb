#### Use esse código para descobrir qual a posição de um item que queira clicar

```python
import pyautogui
import pyperclip
import time
import keyboard

#importar do selenium o webdriver
'''
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
#criar o navegador
#navegador = webdriver.Chrome("chromedriver.exe")
'''

pyautogui.PAUSE = 1

while True:

    time.sleep(2)
    #clicar na aba da ASC - passo 0
    pyautogui.click(x=1283, y=0)

    # digitar a opção 2 - passo 1
    pyautogui.click(x=1667, y=976)
    pyautogui.press("o")
    pyautogui.press("enter")

    #aguarda após o oi - passo 2
    time.sleep(8)
    pyautogui.press("2")
    pyautogui.press("enter")

    #clicar na planilha intenções - passo 3
    time.sleep(2)
    pyautogui.click(x=1117, y=0)

    # setinha pra baixo no teclado e ctrl + c - passos 4 e 5
    pyautogui.press("down")
    pyautogui.hotkey("ctrl","c")
    time.sleep(1)

    # clica na aba da ASC de novo e tecla ctrl + v (clica no chat) - Passo 6
    pyautogui.click(x=1283, y=0)
    time.sleep(1)
    pyautogui.click(x=1667, y=976)
    pyautogui.hotkey("ctrl","v")
    pyautogui.press("enter")
    time.sleep(1)


    #clicar 3x para selecionar a resposta (position) - passo 7
    pyautogui.click(x=1633, y=890, clicks=3)
    pyautogui.click(x=1633, y=890, button='right')
    pyautogui.click(x=1611, y=738)

    #volta pra planilha de intenções
    pyautogui.click(x=1117, y=0)

    #na planilha, ctrl + v
    pyautogui.press("right")
    pyautogui.press("right")
    pyautogui.press("enter")
    pyautogui.hotkey("ctrl","v")

    #enter de novo e voltar ao ponto inicial
    pyautogui.press("enter")
    pyautogui.press("up")
    pyautogui.press("left")
    pyautogui.press("left")

    #voltar para a tela da asc
    pyautogui.click(x=1283, y=0)

    #clica em reiniciar
    pyautogui.click(x=1818, y=983)
    time.sleep(12)   
    pyautogui.click(x=1859, y=984)


if pyautogui.press('b'):
    break
else:
    start()
   


#fazer o loop



'''
pyautogui.hotkey("ctrl","t")
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
'''

time.sleep(3)

pyautogui.position()
```
